```python
miss_per = df.isnull().sum()/df.shape[0] * 100
miss_per
```

```python
df2 = df.drop(columns=miss_cols)
```

```python
# graph for these datatypes (heatmap)
df3 = df2.select_dtypes(include=["int64", "float64"])
plt.figure(figsize=(25, 25))
sns.heatmap(df3.isnull())
```

```python
df3.head()
```

```python
df3[df3.isnull().any(axis=1)]
```

```python
df3.isnull().sum()
```

```python
# to get variables with missing values
miss_num_var = [var for var in df3.columns if df3[var].isnull().sum() > 0]
miss_num_var
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(miss_num_var):
    plt.subplot(2, 2, i+1)
    sns.distplot(df3[var], bins=20, kde_kws={"linewidth":5, "color":"#DC143C"})
```

In the Above distplot,

1st graph, we can see that the graph is symmetric, we can also call it as normal distribution where we can either apply mean, median or mode to fill the missing data

In 2nd and 3rd graph, we can see that the graph is skewed, ie. unsymmentric, we can also say that the graph is not normally distributed,
then at those places we need to apply median to fill those missing data

```python
df4_mean = df3.fillna(df3.mean())
```

```python
df4_mean.isnull().sum().sum() # now we have 0 missing data.
```

```python
df4_mean.head()
```

```python
miss_num_var
```

```python
df4_mean['LotFrontage']
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(miss_num_var):
    plt.subplot(2, 2, i+1)
    sns.distplot(df3[var], bins=20, kde_kws={"linewidth":8, 'color':'red'}, label="Original",)
    sns.distplot(df4_mean[var], bins=20, kde_kws={"linewidth":8, 'color':'green'}, label="Mean",)
    plt.legend()
```

```python
df5_median = df3.fillna(df3.median())
```

```python
df4_mean.isnull().sum().sum() # now we have 0 missing data.
```

```python
df5_median.head()
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(miss_num_var):
    plt.subplot(2, 2, i+1)
    sns.distplot(df3[var], bins=20, kde_kws={"linewidth":8, 'color':'red'}, label="Original",)
    sns.distplot(df4_mean[var], bins=20, kde_kws={"linewidth":8, 'color':'green'}, label="Mean",)
    sns.distplot(df5_median[var], bins=20, kde_kws={"linewidth":8, 'color':'k'}, label="Median",)
    plt.legend()
```

```python
for i, var in enumerate(miss_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_median[var])
```

```python
df_concat = pd.concat([df3[miss_num_var],df4_mean[miss_num_var],df5_median[miss_num_var]], axis=1)
```
